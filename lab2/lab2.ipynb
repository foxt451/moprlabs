```python
import math

def f1(x1, x2):
  return (x1-x2)**2 + 1/9*(x1 + x2 - 10)**2

def f2(x1, x2):
  return 100*(x2 - x1**2)**2 + (1-x1)**2

def f3(x1, x2):
  return (math.sin(x1-0.6) - x2 - 1.6)**2 + (3*x1 - math.cos(x2) - 0.9)**2
```

```python
def central_diff(func, h, *args):
  grad = []
  for i in range(len(args)):
    args_copy = list(args)
    args_copy[i] += h
    f1 = func(*args_copy)
    args_copy[i] -= 2 * h
    f2 = func(*args_copy)
    grad.append((f1 - f2) / (2 * h))
  return grad
```

```python
import pandas as pd
diffs = [10 ** i for i in range(-2, -10, -1)]
data = [[f1, 0, 1], [f2, -1.2, 1], [f3, 0, 0]]
display(diffs)
df = pd.DataFrame([[central_diff(func, h, *args) for h in diffs] for [func, *args] in data])
pd.options.display.float_format = '{:,}'.format

display(df)

def flatten_columns(df, cols):
    """Flattens multiple columns in a data frame"""
    flattened_cols = {}
    for col in cols:
        flattened_cols[col] = pd.DataFrame([value for (index, values) in df[col].items() for value in values])
    flattened_df = pd.concat([flattened_cols[col] for col in cols], axis=1)
    return flattened_df
df = flatten_columns(df, df.columns)
df.index = pd.MultiIndex.from_tuples(list(zip([item for func in ['F1', 'F2', 'F3'] for item in [func, func]], ['x1', 'x2'] * 3)))
df.columns = pd.Index([diff for diff in diffs], name='h')
df.transpose()
```
